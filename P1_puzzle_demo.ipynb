 # Demonstration of the puzzle dataset
 The dirctory is structured as follows:
 - `data`
     - `size-100`
         - `raster`
             - `0.png`
             - `...`
             - `108.png`
         - `pieces.json`
     - `size-500`
         - `raster`
             - `0.png`
             - `...`
             - `520.png`
         - `pieces.json`

 Importantly, you have the `raster` directory with the puzzle pieces and the `pieces.json` file with the piece information.
 The `pieces.json` is a dictionary with the piece id as key and for each piece the following information:
 - `x1`: x coordinate of the top left corner
 - `y1`: y coordinate of the top left corner
 - `x2`: x coordinate of the bottom right corner
 - `y2`: y coordinate of the bottom right corner

 We import some packages to load an examplary puzzle image and the piece information from thejson file.
